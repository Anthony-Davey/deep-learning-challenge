```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
```

```python
data = pd.read_csv("./charity_data.csv")
```

```python
data
```

```python
data = data.drop(columns = ['EIN', 'NAME'])
```

```python
data.nunique()
```

```python
binning_counts = data['APPLICATION_TYPE'].value_counts()
```

```python

application_type_replace = list(binning_counts[binning_counts<300].index)
application_type_replace
```

```python
for a_type in application_type_replace:
  data['APPLICATION_TYPE'] = data['APPLICATION_TYPE'].replace(a_type,"Other")

data['APPLICATION_TYPE'].value_counts()
```

```python
counts_binning = data['CLASSIFICATION'].value_counts()
```

```python
class_count = counts_binning[counts_binning>1]
class_count
```

```python
classifications_to_replace = list (counts_binning[counts_binning<100].index)
classifications_to_replace
```

```python
for cls in classifications_to_replace:
  data['CLASSIFICATION'] = data['CLASSIFICATION'].replace(cls,"Other")

data['CLASSIFICATION'].value_counts()
```

```python
data = pd.get_dummies(data, dtype=float)
data.head()
```

```python
y = data['IS_SUCCESSFUL']
y
```

```python
X = data.drop('IS_SUCCESSFUL', axis = 1)
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=3)
```

```python
scaler = StandardScaler()

X_scaler = scaler.fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
input_features = len( X_train_scaled[0])
hidden_nodes_layer1 = 21
hidden_nodes_layer2 = 14
hidden_nodes_layer3 = 7
nn = tf.keras.models.Sequential()

nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=input_features, activation='relu'))
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, input_dim=input_features, activation='relu'))

nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

nn.summary()
```

```python
nn.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
fit_model = nn.fit(X_train_scaled, y_train,epochs=100)
```

```python
model_loss, model_accuracy = nn.evaluate(X_test_scaled, y_test, verbose = 2)
print(f'Loss: {model_loss}, Accuracy: {model_accuracy}')
```
